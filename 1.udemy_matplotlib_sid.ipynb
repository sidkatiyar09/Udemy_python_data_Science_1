# OVERLAPping LEGENDS AND GRAPH

````python
fig,ax= plt.subplots()

ax.plot(x,x**2,label = 'square legend')
ax.plot(x,x**3,label = 'quad legend')
ax.plot(x,-y,label = 'inverted legend')

ax.set_xlabel('X- Label')
ax.set_ylabel('Y-Label')
ax.set_title('TITLE')

ax.legend(loc=4)  # loc = 4 is right lower position of legend
````

# **PLOT APPEARANCE**

````python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(5,2),dpi=150)
ax[0].plot(x,y,color='cyan')
ax[1].plot(y,x**5,color='green')
plt.tight_layout()
````

````python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,2),dpi=150)
ax[0].plot(x,y,color='purple',linewidth=4,linestyle='--')
ax[1].plot(y,x**5,color='green',linewidth=2,alpha=.5,linestyle='-.')
plt.tight_layout()
````

````python
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(8,5),dpi=150)
ax.plot(x,y,color='purple',linewidth=2,linestyle='--',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')

plt.tight_layout()
````

#**x-ticks & y-ticks**

````python
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(8,5),dpi=150)
ax.plot(x,y,color='purple',linewidth=2,linestyle='--',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')

ax.set_xticks([1,2,3,4,5,6,7])
ax.set_yticks([0,7,12,22,30])
plt.tight_layout()
````
